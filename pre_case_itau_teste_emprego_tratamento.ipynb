```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('/content/drive/MyDrive/pre-case-itau/emprego.csv',
                    sep=';', encoding='UTF-8')
dados.head()
```

```python
dados = dados.rename(columns={'ID Emprego': 'id_emprego', 'ID Cliente': 'id_cliente',
                              'Tipo de Emprego': 'tipo_emprego'})
```

```python
dados.dtypes
```

```python
dados['tipo_emprego'] = dados.tipo_emprego.str.upper()
```

```python
import unicodedata
def accent_remover(text):
  return ''.join(caracter for caracter in unicodedata.normalize('NFD', text)
  if unicodedata.category(caracter) != 'Mn')
```

```python
dados['tipo_emprego'] = dados['tipo_emprego'].apply(accent_remover)
dados.head()
```

```python
dados.to_csv('/content/drive/MyDrive/pre-case-itau/emprego_tratado.csv', sep=';', encoding='UTF-8', index=False)
```
