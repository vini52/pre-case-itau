```python
import pandas as pd
import numpy as np
```

```python
df_cliente = pd.read_csv('/content/drive/MyDrive/pre-case-itau/cliente_tratado.csv',
                         sep=';', encoding='UTF-8')
```

```python
df_emprestimo = pd.read_csv('/content/drive/MyDrive/pre-case-itau/emprestimo_tratado.csv',
                         sep=';', encoding='UTF-8')
```

```python
df_residencia = pd.read_csv('/content/drive/MyDrive/pre-case-itau/residencia_tratado.csv',
                         sep=';', encoding='UTF-8')
```

```python
df_emprego = pd.read_csv('/content/drive/MyDrive/pre-case-itau/emprego_tratado.csv',
                         sep=';', encoding='UTF-8')
```

```python
df_pagamento = pd.read_csv('/content/drive/MyDrive/pre-case-itau/pagamento_tratado.csv',
                         sep=';', encoding='UTF-8')
```

```python
df_cliente.loc[df_cliente['id_cliente'] == 4, 'tipo_emprego'] = 'DESEMPREGADO'
```

```python
df_cliente.head()
```

```python
df_pagamento['status_pagamento'] = df_pagamento['status_pagamento'].replace({'Pago': True, 'Pendente': False}).astype(bool)
```

```python
df_pagamento.head()
```

```python
df_emprestimo['status_pagamento'] = df_emprestimo['status_pagamento'].replace({'PAGO': True, 'EM ANDAMENTO': False, 'ATRASADO': False}).astype(bool)
```

```python
df_emprestimo.head()
```

```python
dados_combinados = pd.merge(df_cliente, df_emprestimo, on='id_cliente', how='inner')
dados_combinados = pd.merge(dados_combinados, df_residencia, on='id_cliente', how='inner')
dados_combinados = pd.merge(dados_combinados, df_emprego, on='id_cliente', how='inner')
dados_combinados = pd.merge(dados_combinados, df_pagamento, on='id_emprestimo', how='inner')
dados_combinados.drop(columns=['tipo_emprego_y', 'status_pagamento_y'], inplace=True)
```

```python
dados_combinados.head()
```

```python
dados_combinados.dtypes
```

```python
dados_combinados.rename(columns={'tipo_emprego_x': 'tipo_emprego', 'status_pagamento_x': 'status_pagamento'}, inplace=True)
```

```python
dados_combinados.head()
```

```python
dados_combinados.dtypes
```

```python
dados_combinados['renda'] = dados_combinados['renda'] * 1000
dados_combinados['renda'].head()
```

```python
dados_combinados['valor'] = dados_combinados['valor'] * 1000
dados_combinados['valor'].head()
```

```python
# Calcular estatsticas descritivas para valor do emprstimo por tipo de emprstimo
desc_stats = dados_combinados.groupby('tipo_emprestimo')['valor'].describe()
print(desc_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotar grfico de barras para tipo de emprstimo vs valor do emprstimo
plt.figure(figsize=(10, 6))
sns.barplot(data=dados_combinados, x='tipo_emprestimo', y='valor')
plt.title('Valor do Emprstimo por Tipo de Emprstimo')
plt.xlabel('Tipo de Emprstimo')
plt.ylabel('Valor do Emprstimo')
plt.xticks(rotation=45)
plt.show()
```

```python
# Criar tabela de contingncia entre tipo de emprstimo e tipo de residncia
contingency_table = pd.crosstab(dados_combinados['tipo_emprestimo'], dados_combinados['residencia_permanente'])

# Selecionar apenas os dados para emprstimos habitacionais
habitacional_counts = contingency_table.loc['HABITACIONAL']

# Exibir contagem de ocorrncias para emprstimos habitacionais por tipo de residncia
print(habitacional_counts)
```

```python
# Plotar grfico de barras para propores de emprstimos habitacionais por tipo de residncia
plt.figure(figsize=(8, 6))
habitacional_counts.plot(kind='bar')
plt.title('Emprstimos Habitacionais por Tipo de Residncia')
plt.xlabel('Tipo de Residncia')
plt.ylabel('Nmero de Emprstimos')
plt.xticks(rotation=45)
plt.show()
```

```python
dados_combinados = dados_combinados.drop(columns=['tipo_residencia'])
```

```python
dados_combinados.drop(columns=['nome', 'nome_mae', 'cidade', 'estado', 'endereco'], inplace=True)
```
