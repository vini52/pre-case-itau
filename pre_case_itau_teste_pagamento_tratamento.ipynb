```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('/content/drive/MyDrive/pre-case-itau/pagamento.csv',
                    sep=';', encoding='UTF-8')
dados.head()
```

```python
dados = dados.rename(columns={'ID Status Pagamento': 'id_status_pagamento', 'ID Empr√©stimo': 'id_emprestimo',
                              'Data de Pagamento': 'data_pagamento', 'Valor Pago': 'valor_pago', 'Status':'status_pagamento'})
dados.head()
```

```python
dados.dtypes
```

```python
from decimal import Decimal
def converter_valor(valor_moeda):
  valor_formatado = valor_moeda.replace('R$', '').replace(' ', '')
  return Decimal(valor_formatado)
dados['valor_pago'] = dados['valor_pago'].apply(converter_valor)
dados.head()
```

```python
from datetime import datetime
def converter_data(data_string):
    return datetime.strptime(data_string, '%d/%m/%Y')
dados['data_pagamento'] = dados['data_pagamento'].apply(converter_data)
dados.head()
```

```python
dados.dtypes
```

```python
dados['status_pagamento'] = dados['status_pagamento'].replace({'Pago': True,'Pendente': False}).astype(bool)
dados.head()
```

```python
dados.dtypes
```

```python
dados.to_csv('/content/drive/MyDrive/pre-case-itau/pagamento_tratado.csv', sep=';', encoding='UTF-8', index=False)
```
