```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('/content/drive/MyDrive/pre-case-itau/cliente.csv',
                    sep=';', encoding='UTF-8')
```

```python
dados = dados.rename(columns={'Tipo de Emprego': 'emprego'})
```

```python
dados.head()
```

```python
dados = dados.rename(columns={'Nome': 'nome', 'Idade': 'idade', 'Renda Mensal': 'renda', 'Nome da Mãe': 'nome_mae',
                              'Residência Permanente': 'residencia_permanente', 'Cidade': 'cidade', 'Estado': 'estado'})
```

```python
dados.head()
```

```python
dados = dados.rename(columns={'ID Cliente': 'id_cliente'})
```

```python
dados.head()
```

```python
dados = dados.rename(columns={'emprego': 'tipo_emprego'})
```

```python
dados.dtypes
```

```python
from decimal import Decimal
def converter_valor(valor_moeda):
  valor_formatado = valor_moeda.replace('R$', '').replace(' ', '')
  return Decimal(valor_formatado)
dados['renda'] = dados['renda'].apply(converter_valor)
dados.head()
```

```python
dados.dtypes
```

```python
dados['nome_mae'] = dados.nome_mae.str.upper()
dados.head()
```

```python
import unicodedata
def accent_remover(text):
  return ''.join(caracter for caracter in unicodedata.normalize('NFD', text)
  if unicodedata.category(caracter) != 'Mn')
```

```python
dados['cidade'] = dados['cidade'].apply(accent_remover)
```

```python
dados.head()
```

```python
dados['residencia_permanente'] = dados['residencia_permanente'].replace({'SIM': True, 'NÃO': False}).astype(bool)
```

```python
dados.head()
```

```python
dados['tipo_emprego'] = dados['tipo_emprego'].apply(accent_remover)
dados['tipo_emprego'] = dados.tipo_emprego.str.upper()
dados.head()
```

```python
dados['cidade'] = dados.cidade.str.upper()
dados.head()
```

```python
dados.dtypes
```

```python
dados.to_csv('/content/drive/MyDrive/pre-case-itau/cliente_tratado.csv', sep=';', encoding='UTF-8', index=False)
```
